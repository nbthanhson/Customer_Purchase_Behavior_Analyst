```python
def calc_null_rate(df):
    new_df = df.isnull().sum().to_frame('null_count')
    new_df[['null_rate']] = new_df[['null_count']] / len(df)
    return new_df.sort_values(by = ['null_count'], ascending = False)
```

```python
calc_null_rate(df_fact)
```
